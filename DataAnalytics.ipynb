```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/sample_data/MS_Dhoni_ODI_record.csv')

#Basic Check
print(df.head())
print(df.tail())

# replace v in opposition  [Data cleaning ]
df['opposition'] = df['opposition'].str.replace('v','',regex=False) # method1
# df['opposition'] =df['opposition'].apply(lambda x: x[2:])  # method2


df['date'] = pd.to_datetime(df['date'], dayfirst= True)
df['Year'] = df['date'].dt.year.astype(int)

# df['date'] = pd.to_datetime(df['date'])
# batting avg = total runs scored /(Innigs count - not outs count)

df['score'] = df['score'].apply(str)
df['not_out'] = np.where(df['score'].str.endswith('*'),1,0)
df.drop(columns='odi_number',inplace=True)

#droping those innings where Dhoni did not bat and storing in a new DataFrame
#Take all the columns , start with runs_scored
df_new = df.loc[((df['score'] != 'DNB')& (df['score'] != 'TDNB')), 'runs_scored':]
print(df_new.head())

#Defining proper Datatypes...
df_new['runs_scored']= df_new['runs_scored'].astype(int)
df_new['balls_faced']= df_new['balls_faced'].astype(int)
df_new['strike_rate']= df_new['strike_rate'].astype(float)
df_new['fours']= df_new['fours'].astype(int)
df_new['sixes']= df_new['sixes'].astype(int)

#Career Stat
first_match_date = df['date'].dt.date.min().strftime('%B %d, %Y')    # First Match    # %B is use to give us month
print('First match date : ',first_match_date)

last_match_date = df['date'].dt.date.max().strftime('%B %d, %Y')    # last Match    # %B is use to give us month
print('Last match date : ',last_match_date)

number_of_matches = df.shape[0]  # number of matches played in career
print("Number of matches played : ",number_of_matches)
number_of_inns = df_new.shape[0]  # number of innings
print("Number of innings played : ",number_of_inns)

ball_faced = df_new['balls_faced'].sum()
runs_scored = df_new['runs_scored'].sum()    #runs scored in carrer
print("Run scored in carrer is : ",runs_scored)
print("Ball faced in carrer is : ",ball_faced)

carrer_sr = (runs_scored/ball_faced)*100  # career strike rate
print('Career Strike rate is : {:.2f}'.format(carrer_sr))

hundreds = (df_new['runs_scored'] >=100).sum()
print('Number of 100s: ', hundreds)

fifties = ((df_new['runs_scored']>=50) & (df_new['runs_scored'] <100)).sum()
print('Number of Fifties : ',fifties)

#number of matches played against different opposition
#Counts the occurences of each unique values in the 'opposition' columns
#opposition_count will be a series with a labeled index as opposition
opposition_count= df['opposition'].value_counts()
print(opposition_count)
#Plot the count as a bar plot
opposition_count.plot(kind='bar',title = "Number of matches", figsize=(8,5))
plt.show()
```

```python
#Runs scored against each team
#Group the DataFrame by 'opposition' column
grouped_by_opposition = df_new.groupby('opposition')
#Sum the 'runs_score' column for each group
sum_of_run_scored = grouped_by_opposition['runs_scored'].sum()
print(sum_of_run_scored)
#Sum_of_scored is a series with a labelled index , which is opposition
#Convert it into a DataFrame and remove the index
runs_scored_by_opposition = pd.DataFrame(sum_of_run_scored).reset_index()
runs_scored_by_opposition.plot(x='opposition', kind='bar', title ='Runs scored',figsize=(8,6))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats

#Generate random data with a larger sample size
sample_size = 100000
rating = np.random.randint(1, 6, size=sample_size)

#modifiy data for the right skew
right_skew_data = np.concatenate((rating, np.random.randint(1,2,sample_size//2)))

#modify data for the left skew
left_skew_data = np.concatenate((rating,np.random.randint(4,6,sample_size//2)))


#Function to calculate mean, mode, median and skewness
def calculate_stats(data):
    mean = np.mean(data)
    mode = stats.mode(data)
    median = np.median(data)
    skewness = stats.skew(data)
    return mean, mode, median, skewness

#Function to plot histogram with mean and median lines
def plot_histogram(ax, data, title):
    ax.hist(data, bins=np.arange(1,7)-0.5, rwidth=0.8, alpha=0.7)
    ax.set_title(title)
    ax.set_xlabel('Rating')
    ax.set_ylabel('Frequency')
    ax.set_xticks(np.arange(1,6))
    ax.grid(axis='y', alpha=0.5)

    #calculate mean, median and skewness
    mean, mode, median, skewness = calculate_stats(data)
    ax.axvline(x=mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
    ax.axvline(x=median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
    ax.legend()

    #print mean, median and skewness
    print(f"{title}:")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Skewness: {skewness:.2f}")
    print()

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
plot_histogram(axs[1], right_skew_data, "Right Skewed Data")
plot_histogram(axs[2], left_skew_data, "Left Skewed Data")
plot_histogram(axs[0], rating, "Original Data")
plt.tight_layout()
plt.show()
```

```python
#Calculate mean and standard deviation of the glucose column
glucose_mean= df['Glucose'].mean()
glucose_std = df['Glucose'].std()

print(f'Mean of Glucose: {glucose_mean}')
print(f'Standard Deviation of Glucose: {glucose_std}')

#Calculate  Z score for glucose level of 168(using manual method)
glucose_value= 168
z_score_168_manual= (glucose_value - glucose_mean) / glucose_std
print(f'Z score for Glucose level of 168: {z_score_168_manual}')

#calculate the z score using zscore function
#This will add a column for the zscore of glucose for each record
df['Glucose_zscore'] = zscore(df['Glucose'])
print(df)


#Find the row index where glucose =168
#df['Glucose']==glucose_value = Retrun "True" where Glucose =168
#df[now gets the actual records for 168]
#.index return the indices of row matching this condition
row_index= df[df['Glucose'] == 168].index
print(f'Row index where Glucose = 168: {row_index}')

#Now we extract the first row from Glucose_zscore DF where Glucose =168
#Hence row_index[0]... Remember, we may have multiple rows matching 168
z_score_168_using_function = df['Glucose_zscore'].iloc[row_index[0]]

print(f'Z score for Glucose level of 168: {z_score_168_using_function}')

#Calculate the percentile for this z score
percentile_168 = stats.norm.cdf(z_score_168_manual)*100    #Multiply by 100 to gert percentage
print(f'Percentile for Glucose level of 168: {percentile_168}%')

```

```python
#Find the z-Score at the 30th percentile
z_score_30 = stats.norm.ppf(0.30)
print(f'Z score at the 30th percentile: {z_score_30}')

#Calculate the corresponding Glucose value for the z - score of the 30th percentile
glucose_value_30 = z_score_30 * glucose_std + glucose_mean
print(f'Glucose value at the 30th percentile: {glucose_value_30}')

```

```python
#Given z-score of 2.8, calculate the corresponding Glucose level
z_score_given = 2.8
glucose_value_given = z_score_given * glucose_std + glucose_mean
print(f'Glucose value for z-score of 2.8: {glucose_value_given}')

#Calculate the percentile for the z-score of 2.8
percentile_given = stats.norm.cdf(z_score_given)*100           #Multiply by 100 to get percentage
print(f'Percentile for z-score of 2.8: {percentile_given}%')
```

```python
df = pd.read_csv('/content/sample_data/tips.csv')
```

```python
#Extract total_bill column into a new Series
total_bill_series = df['total_bill']
print(total_bill_series.head())


#Calculate mean and standard deviation
mu, std = total_bill_series.mean(), total_bill_series.std()
print(f'Mean: {mu}')
print(f'Standard Deviation: {std}')


#Calculate z-score for each value
z_scores = (total_bill_series - mu) / std

#Plot normal distrbution graph of total_bill
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.hist(total_bill_series, bins=20, density=True, alpha=0.6, color='b', edgecolor='black')

x= np.linspace(min(total_bill_series), max(total_bill_series), 100)
p= stats.norm.pdf(x,mu,std)
plt.plot(x,p,'k',linewidth=2)
title = "Fit result: mu = %.2f, sigma = %.2f" % (mu, std)
plt.title(title)
plt.xlabel('Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.grid(True)

#QQ Plot
plt.subplot(1,2,2)
stats.probplot(total_bill_series, dist='norm', plot=plt)
plt.title('QQ Plot of Total Bill')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Plot a graph showing mean +- 3 SD
plt.figure(figsize=(10,6))
plt.hist(total_bill_series, bins=20, density=True, alpha=0.6, color='b', edgecolor='black')

#Fit a normal distribution to the data
x = np.linspace(min(total_bill_series), max(total_bill_series), 1000)
p = stats.norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit result: mu = %.2f, sigma = %.2f" % (mu, std)
plt.title(title)
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.grid(True)

#Plot mean +- 3 SD
for i in range(1,4):
    plt.axvline(mu + i*std, color='g', linestyle='--', linewidth = 2)
    plt.axvline(mu - i*std, color='r', linestyle='--', linewidth = 2, label=f'Mean - {i} SD')
plt.legend()
plt.show()
```

```python
#Check if mean +- 1SD , +-2SD, and mean +-3SD satisfy the empirical rule
within_1SD = (z_scores <= 1) & (z_scores >= -1)
within_2SD = (z_scores <= 2) & (z_scores >= -2)
within_3SD = (z_scores <= 3) & (z_scores >= -3)

percentage_within_1SD = np.sum(within_1SD) / len(total_bill_series) * 100
percentage_within_2SD = np.sum(within_2SD) / len(total_bill_series) * 100
percentage_within_3SD = np.sum(within_3SD) / len(total_bill_series) * 100

print(f"Percentage of data within mean +- 1 SD : {percentage_within_1SD: .2f}%")
print(f"Percentage of data within mean +- 2 SD : {percentage_within_2SD: .2f}%")
print(f"Percentage of data within mean +- 3 SD : {percentage_within_3SD: .2f}%")
```

```python
#Create a new dataframe to store total_bill , z_score, and within_1SD , within_2SD, within_3SD flags
result_df = pd.DataFrame({
    'total_bill': total_bill_series,
    'z_score': z_scores,
    'within_1SD': within_1SD,
    'within_2SD': within_2SD,
    'within_3SD': within_3SD
})

print(result_df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import shapiro
```

```python
df = pd.read_csv('/content/sample_data/diabetes.csv')
df.head()
```

```python
#Shapiro test on Glucose Columns

shapiro_test = stats.shapiro(df['Glucose'])
print('Shapiro-Wilk p-value :', shapiro_test.pvalue)


if shapiro_test.pvalue > 0.05:
  print('The data is normally distributed')
else:
  print('The data is not normally distributed, but the q-q plot can provide further insights')

#Generate a Q-Q plot for the sample mean
stats.probplot(df['Glucose'], dist='norm', plot=plt)
plt.title('Q-Q plot for Glucose Column')
plt.show()


#Central Limit Theorem(CLT)
sample_mean=[]
n_sample = 100
sample_size = 30

for _ in range(n_sample):
  sample = df['Glucose'].sample(n=sample_size, replace=True)
  sample_mean.append(sample.mean())

#plot the distribution of sample mean
plt.hist(sample_mean, bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Sample Mean')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()


#Generate a Q-Q Plot for the sample mean
stats.probplot(sample_mean, dist='norm', plot =plt)
plt.title('Q-Q Plot of sample means')
plt.show()


#Perform the shapiro-wilk test on the sample means
shapiro_test = stats.shapiro(sample_mean)
print('Shapiro -Wilk- p_value : ', shapiro_test.pvalue)

#Interpret the result
if shapiro_test.pvalue < 0.05:
    print('The data likely does not follow a normal distribution .')
else :
    print('The data may be normally distributed , but the q-q plot can provide further insights ')
```

```python
import pandas as pd
from scipy import stats
```

```python
data = pd.read_csv('/content/sample_data/Mall_Customers.csv')
age_data = data['Age']
```

```python
#Hypothesize population
pop_avg_age = 40

#perform one -sample test
t_statistic, p_value = stats.ttest_1samp(age_data, pop_avg_age)
dof = len(age_data)-1


print("T statistics : ", t_statistic)
print("P value : ", p_value)
print("Degree of freedom : ", dof)
```

```python
#interpretation
alpha = 0.025
if p_value < alpha:
  print("Reject the (mean age = 40) null hypothesis")
else:
  print("Fail to reject (mean age = 40) the null hypothesis")

critical_value = stats.t.ppf(1-alpha, dof)
print("Critical value : ", critical_value)
```

```python
#Two - Tailed t-test (Because we are just checking if sample mean = population mean)
# So we need to take absolute value of t-statictics
if (abs(t_statistic) > critical_value):
  print("Reject the (mean age = 40) null hypothesis")
else:
  print("Fail to reject (mean age = 40) the null hypothesis")
```

```python
# Separate data by genre
male_spending = data[data['Genre'] == 'Male']['Spending Score (1-100)']
female_spending = data[data['Genre'] == 'Female']['Spending Score (1-100)']

#perform two -sample t-test
t_statistic, p_value = stats.ttest_ind(male_spending, female_spending)

n_male = len(male_spending)
n_female = len(female_spending)
dof = n_male + n_female - 2

alpha = 0.025

#Find the critical value
critical_value = stats.t.ppf(1-alpha, dof)

#print the result
print("T statistics : ", t_statistic)
print("P value : ", p_value)
print("Degree of freedom : ", dof)
print("Critical value : ", critical_value)

if p_value < alpha:
  print("There is a significant diffrence in spending score between male and female groups.")
else:
  print("There is no a significant diffrence in spending score between male and female groups.")

if (abs(t_statistic) > critical_value):
  print("There is a significant diffrence in spending score between male and female groups.")
else:
  print("There is no a significant diffrence in spending score between male and female groups.")



```
