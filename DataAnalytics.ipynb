```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/sample_data/MS_Dhoni_ODI_record.csv')

#Basic Check
print(df.head())
print(df.tail())

# replace v in opposition  [Data cleaning ]
df['opposition'] = df['opposition'].str.replace('v','',regex=False) # method1
# df['opposition'] =df['opposition'].apply(lambda x: x[2:])  # method2


df['date'] = pd.to_datetime(df['date'], dayfirst= True)
df['Year'] = df['date'].dt.year.astype(int)

# df['date'] = pd.to_datetime(df['date'])
# batting avg = total runs scored /(Innigs count - not outs count)

df['score'] = df['score'].apply(str)
df['not_out'] = np.where(df['score'].str.endswith('*'),1,0)
df.drop(columns='odi_number',inplace=True)

#droping those innings where Dhoni did not bat and storing in a new DataFrame
#Take all the columns , start with runs_scored
df_new = df.loc[((df['score'] != 'DNB')& (df['score'] != 'TDNB')), 'runs_scored':]
print(df_new.head())

#Defining proper Datatypes...
df_new['runs_scored']= df_new['runs_scored'].astype(int)
df_new['balls_faced']= df_new['balls_faced'].astype(int)
df_new['strike_rate']= df_new['strike_rate'].astype(float)
df_new['fours']= df_new['fours'].astype(int)
df_new['sixes']= df_new['sixes'].astype(int)

#Career Stat
first_match_date = df['date'].dt.date.min().strftime('%B %d, %Y')    # First Match    # %B is use to give us month
print('First match date : ',first_match_date)

last_match_date = df['date'].dt.date.max().strftime('%B %d, %Y')    # last Match    # %B is use to give us month
print('Last match date : ',last_match_date)

number_of_matches = df.shape[0]  # number of matches played in career
print("Number of matches played : ",number_of_matches)
number_of_inns = df_new.shape[0]  # number of innings
print("Number of innings played : ",number_of_inns)

ball_faced = df_new['balls_faced'].sum()
runs_scored = df_new['runs_scored'].sum()    #runs scored in carrer
print("Run scored in carrer is : ",runs_scored)
print("Ball faced in carrer is : ",ball_faced)

carrer_sr = (runs_scored/ball_faced)*100  # career strike rate
print('Career Strike rate is : {:.2f}'.format(carrer_sr))

hundreds = (df_new['runs_scored'] >=100).sum()
print('Number of 100s: ', hundreds)

fifties = ((df_new['runs_scored']>=50) & (df_new['runs_scored'] <100)).sum()
print('Number of Fifties : ',fifties)

#number of matches played against different opposition
#Counts the occurences of each unique values in the 'opposition' columns
#opposition_count will be a series with a labeled index as opposition
opposition_count= df['opposition'].value_counts()
print(opposition_count)
#Plot the count as a bar plot
opposition_count.plot(kind='bar',title = "Number of matches", figsize=(8,5))
plt.show()
```

```python
#Runs scored against each team
#Group the DataFrame by 'opposition' column
grouped_by_opposition = df_new.groupby('opposition')
#Sum the 'runs_score' column for each group
sum_of_run_scored = grouped_by_opposition['runs_scored'].sum()
print(sum_of_run_scored)
#Sum_of_scored is a series with a labelled index , which is opposition
#Convert it into a DataFrame and remove the index
runs_scored_by_opposition = pd.DataFrame(sum_of_run_scored).reset_index()
runs_scored_by_opposition.plot(x='opposition', kind='bar', title ='Runs scored',figsize=(8,6))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats

#Generate random data with a larger sample size
sample_size = 100000
rating = np.random.randint(1, 6, size=sample_size)

#modifiy data for the right skew
right_skew_data = np.concatenate((rating, np.random.randint(1,2,sample_size//2)))

#modify data for the left skew
left_skew_data = np.concatenate((rating,np.random.randint(4,6,sample_size//2)))


#Function to calculate mean, mode, median and skewness
def calculate_stats(data):
    mean = np.mean(data)
    mode = stats.mode(data)
    median = np.median(data)
    skewness = stats.skew(data)
    return mean, mode, median, skewness

#Function to plot histogram with mean and median lines
def plot_histogram(ax, data, title):
    ax.hist(data, bins=np.arange(1,7)-0.5, rwidth=0.8, alpha=0.7)
    ax.set_title(title)
    ax.set_xlabel('Rating')
    ax.set_ylabel('Frequency')
    ax.set_xticks(np.arange(1,6))
    ax.grid(axis='y', alpha=0.5)

    #calculate mean, median and skewness
    mean, mode, median, skewness = calculate_stats(data)
    ax.axvline(x=mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
    ax.axvline(x=median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
    ax.legend()

    #print mean, median and skewness
    print(f"{title}:")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Skewness: {skewness:.2f}")
    print()

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
plot_histogram(axs[1], right_skew_data, "Right Skewed Data")
plot_histogram(axs[2], left_skew_data, "Left Skewed Data")
plot_histogram(axs[0], rating, "Original Data")
plt.tight_layout()
plt.show()
```
